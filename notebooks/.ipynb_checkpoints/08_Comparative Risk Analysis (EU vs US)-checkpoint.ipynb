# 08 — Comparative Risk Analysis (EU vs US)

This notebook synthesises the results of previous modules to compare the risk–return profile of the IBEX 35 and the S&P 500 over the full sample and across market regimes. This notebook tries to figure out how European and US equity markets differ in terms of volatility, drawdowns, tail risk, and contagion over time

The analysis is built on:
- **Realized returns** (level, volatility, drawdowns).
- **Model-implied risk** (EGARCH volatility, VaR/ES backtests).
- **Relative and dynamic measures** (volatility ratios, rolling beta IBEX|SPX, rolling correlation).

The goal is to produce a concise, portfolio-oriented comparison that can be consumed directly in a dashboard Power BI presentation.

### Objectives

- Quantify and compare realised volatility in IBEX vs S&P 500.
- Measure relative risk via volatility ratios and worst drawdowns.
- Analyze time-varying dependence through rolling correlations and beta IBEX|SPX.
- Contrast tail risk using VaR / ES exceedance statistics by region.
- Derive a compact EU vs US risk profile suitable for reporting and asset-allocation discussion.
-

### Step 1: Setup & Data Loading

This step prepares the working environment and imports all datasets required for the EU–US comparative risk analysis.

```python
# STEP 1 SETUP & DATA LOADING 

import pandas as pd
import numpy as np
from pathlib import Path

# 1.0 Define project paths
PROJECT_ROOT  = Path.cwd().parent                 # notebooks/ → root/
DATA_PATH     = PROJECT_ROOT / "data"
FEATURES_PATH = DATA_PATH / "features"
RESULTS_PATH  = DATA_PATH / "results"

# Save in folders
FIGURES_PATH  = PROJECT_ROOT / "figures"
RISK8_DIR = FIGURES_PATH / "08_comparative_risk"
RISK8_DIR.mkdir(parents=True, exist_ok=True)

print("Project paths loaded:")
for p in [DATA_PATH, FEATURES_PATH, RESULTS_PATH, FIGURES_PATH]:
    print(" -", p)

# 1.1 Load returns dataset
print("\nLoading returns dataset...")

returns_file = FEATURES_PATH / "returns.csv"
if not returns_file.exists():
    raise FileNotFoundError(f"Returns file not found: {returns_file}")

returns = pd.read_csv(
    returns_file,
    index_col=0,
    parse_dates=True
)

returns.columns = ["IBEX", "SPX"]

print("Returns columns:", list(returns.columns))
print("IBEX:", returns[['IBEX']].dropna().shape)
print("SPX :", returns[['SPX']].dropna().shape)

# 1.2 Load technical indicators (optional)
ti_file = FEATURES_PATH / "technical_indicators.csv"
if ti_file.exists():
    tech_ind = pd.read_csv(ti_file, index_col=0, parse_dates=True)
    print("\nTechnical indicators loaded:", tech_ind.shape)
else:
    tech_ind = None
    print("\nTechnical indicators not found — continuing without them.")

# 1.3 Load EGARCH volatility (CSV)
egarch_file = RESULTS_PATH / "egarch_volatility.csv"

print("\nLoading EGARCH volatility...")

if egarch_file.exists():
    egarch_vol = pd.read_csv(egarch_file, parse_dates=["Date"])
    egarch_vol = egarch_vol.set_index("Date")
    print("EGARCH volatility loaded:", egarch_vol.shape)
else:
    egarch_vol = None
    print("WARNING: egarch_volatility.csv not found.")

# 1.4 Load DAILY VaR/ES results (REPLACES old var_es_results.csv)
var_es_daily_file = RESULTS_PATH / "var_es_results_daily.csv"

print("\nLoading daily VaR/ES dataset...")

if var_es_daily_file.exists():
    var_es_daily = pd.read_csv(var_es_daily_file, parse_dates=["Date"])
    var_es_daily = var_es_daily.sort_values(["Date", "Index"])
    print("Daily VaR/ES loaded:", var_es_daily.shape)
else:
    var_es_daily = None
    print("WARNING: var_es_results_daily.csv not found — VaR comparisons limited.")

# 1.5 Load walk-forward summary (Notebook 7)
wfsum_file = RESULTS_PATH / "walkforward_errors_summary.csv"

if wfsum_file.exists():
    wf_summary = pd.read_csv(wfsum_file)
    print("\nWalk-forward summary loaded:", wf_summary.shape)
else:
    wf_summary = None
    print("\nWarning: walkforward_errors_summary.csv not found.")

# 1.6 Create aligned return panel (RV needs this)
panel = returns.copy().dropna()
print("\nAligned return panel:", panel.shape)
print(panel.head())

print("\nSetup & data loading completed.")

if "Date" in panel.columns:
    panel = panel.drop(columns=["Date"])

panel.index.name = "Date"

```

### Step 2: Rolling Volatility Comparison

This step quantifies and compares the time-varying volatility of the IBEX 35 and the S&P 500. The aim is that both measures (realised and model-implied) are aligned on the same date index and merged into a single panel, so we can later produce two sets of plots in order to make a comparision easily. That establishes the first layer of the EU–US risk comparison by showing how market volatility evolves over time, which market is structurally more volatile, and how realised volatility contrasts with model-implied EGARCH estimates.


```python
# STEP 2 — ROLLING VOLATILITY COMPARISON

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# 2.0 Load EGARCH volatility (if available)

RESULTS_PATH = Path.cwd().parent / "data" / "results"

try:
    egarch_vol = pd.read_csv(RESULTS_PATH / "egarch_volatility.csv", parse_dates=["Date"])
    egarch_vol = egarch_vol.set_index("Date")
    print("EGARCH volatility loaded successfully.")
except FileNotFoundError:
    egarch_vol = None
    print("EGARCH volatility file not found. Continuing with realized volatility only.")

# 2.1 Compute realized volatility (20-day rolling window)

print("Computing realized volatility (20-day rolling window)...")

rv = panel.rolling(window=20).std() * np.sqrt(252)
rv.columns = ["IBEX_RV", "SPX_RV"]

print("Realized volatility computed:", rv.dropna().shape)

# 2.2 Merge EGARCH volatility (if available)

if egarch_vol is not None:
    print("Aligning EGARCH volatility with return panel...")
    egarch_aligned = egarch_vol.reindex(rv.index)

    rv["IBEX_EGARCH"] = egarch_aligned["Vol_IBEX"]
    rv["SPX_EGARCH"]  = egarch_aligned["Vol_SPX"]

    print("EGARCH volatility merged with realized volatility.")
else:
    print("EGARCH volatility unavailable. Proceeding with realized volatility only.")

# 2.3 Plot realized volatility

plt.figure(figsize=(12, 5))
plt.plot(rv["IBEX_RV"], label="IBEX Realized Volatility", linewidth=1)
plt.plot(rv["SPX_RV"], label="SPX Realized Volatility", linewidth=1)
plt.title("Realized Volatility (20-day rolling)")
plt.ylabel("Volatility (annualized)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save figure correctly
plt.gcf().savefig(RISK8_DIR / "01_realized_volatility.png", dpi=300, bbox_inches="tight")

plt.show()

# 2.4 Plot RV vs EGARCH only if EGARCH exists

if egarch_vol is not None:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    axes[0].plot(rv["IBEX_RV"], label="RV", linewidth=1)
    axes[0].plot(rv["IBEX_EGARCH"], label="EGARCH", linewidth=1)
    axes[0].set_title("IBEX 35 — RV vs EGARCH")
    axes[0].legend()
    axes[0].grid(True)

    axes[1].plot(rv["SPX_RV"], label="RV", linewidth=1)
    axes[1].plot(rv["SPX_EGARCH"], label="EGARCH", linewidth=1)
    axes[1].set_title("S&P 500 — RV vs EGARCH")
    axes[1].legend()
    axes[1].grid(True)

    plt.tight_layout()

    fig.savefig(RISK8_DIR / "02_rv_vs_egarch.png", dpi=300, bbox_inches="tight")

    plt.show()

print("Step 2 completed.")
```

### Step 3: Relative Volatility & Drawdowns

In this step, the notebook evaluates how the risk of the IBEX 35 compares to that of the S&P 500 over time. The ratios used show when the European market is more volatile than the U.S. market, and how this relationship shifts across different periods. Also, the cumulative returns are constructed for each index, which allows the calculation of drawdowns. This quantifies the depth and duration of losses each market has suffered historically. Finally, comparative drawdown plots are generated to visually assess which index experiences more severe and persistent declines.

```python
# STEP 3 — RELATIVE VOLATILITY & DRAWDOWNS 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 3.1 Build analysis panel: returns + realized volatility 
print("Building analysis panel with returns and volatility...")

#Starting panel
analysis_panel = panel.join(rv, how="left")

# incorporating egarch if exists
if egarch_vol is not None:
    egarch_aligned = egarch_vol.reindex(analysis_panel.index)
    analysis_panel["IBEX_EGARCH"] = egarch_aligned["Vol_IBEX"]
    analysis_panel["SPX_EGARCH"]  = egarch_aligned["Vol_SPX"]
    print("EGARCH volatility added to analysis panel.")
else:
    print("EGARCH volatility not available.")

print("Analysis panel shape:", analysis_panel.shape)

# 3.2 Volatility ratios (IBEX vs SPX)
print("\nComputing volatility ratios (IBEX vs SPX)...")

analysis_panel["vol_ratio_RV"] = (
    analysis_panel["IBEX_RV"] / analysis_panel["SPX_RV"]
)

if "IBEX_EGARCH" in analysis_panel.columns and "SPX_EGARCH" in analysis_panel.columns:
    analysis_panel["vol_ratio_EGARCH"] = (
        analysis_panel["IBEX_EGARCH"] / analysis_panel["SPX_EGARCH"]
    )
else:
    print("EGARCH volatility ratio unavailable.")

print("Volatility ratio (RV) summary:")
print(analysis_panel["vol_ratio_RV"].dropna().describe())

# 3.3 Plot volatility ratios over time
print("\nPlotting realized volatility ratios...")

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(analysis_panel.index, analysis_panel["vol_ratio_RV"], 
        label="IBEX/SPX Realized Volatility", linewidth=1.2)

ax.axhline(1.0, color="black", linestyle="--", linewidth=1)

ax.set_title("Relative Volatility (IBEX vs SPX) — Realized Volatility (20-day)")
ax.set_ylabel("Volatility ratio")
ax.grid(True)
ax.legend()
plt.tight_layout()
fig.savefig(RISK8_DIR / "03_volatility_ratio_RV.png", dpi=300, bbox_inches="tight")
plt.show()

# Plot EGARCH ratio if exists
if "vol_ratio_EGARCH" in analysis_panel.columns:
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.plot(analysis_panel.index, analysis_panel["vol_ratio_EGARCH"],
            label="IBEX/SPX EGARCH Volatility", linewidth=1.2)
    ax.axhline(1.0, color="black", linestyle="--", linewidth=1)
    ax.set_title("Relative Volatility (IBEX vs SPX) — EGARCH Conditional Volatility")
    ax.set_ylabel("Volatility ratio")
    ax.grid(True)
    ax.legend()
    plt.tight_layout()
    # SAVE FIGURE 2
    fig.savefig(RISK8_DIR / "04_volatility_ratio_EGARCH.png", dpi=300, bbox_inches="tight")
    plt.show()

# 3.4 Cumulative returns and drawdowns (CORREGIDO)
print("\nComputing cumulative returns and drawdowns...")

# Tus retornos se llaman: "IBEX" y "SPX"
rets = analysis_panel[["IBEX", "SPX"]].dropna()

wealth_ibex = (1 + rets["IBEX"]).cumprod()
wealth_spx  = (1 + rets["SPX"]).cumprod()

dd_ibex = wealth_ibex / wealth_ibex.cummax() - 1
dd_spx  = wealth_spx / wealth_spx.cummax() - 1

analysis_panel["DD_IBEX"] = dd_ibex
analysis_panel["DD_SPX"]  = dd_spx

print("Max drawdown IBEX:", dd_ibex.min())
print("Max drawdown SPX :", dd_spx.min())

# 3.5 Plot comparative drawdowns
print("\nPlotting comparative drawdowns...")

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(dd_ibex.index, dd_ibex, label="IBEX Drawdown", linewidth=1.2)
ax.plot(dd_spx.index, dd_spx, label="SPX Drawdown", linewidth=1.2)

ax.set_title("Comparative Drawdowns — IBEX vs SPX")
ax.set_ylabel("Drawdown")
ax.grid(True)
ax.legend()

plt.tight_layout()

# Save
fig.savefig(RISK8_DIR / "05_volatility_ratio_RV.png", dpi=300, bbox_inches="tight")

plt.show()

print("\nStep 3 completed.")
```

### Step 4: Rolling Beta & Correlation

This step measures how the relationship between the IBEX 35 and the S&P 500 evolves over time, focusing on two key dependence metrics: rolling correlation and rolling beta. First, a clean return panel is prepared using daily IBEX and SPX returns. A 60-day rolling correlation is then computed to capture periods of contagion, decoupling, and structural shifts in cross-market co-movement.

Next, a 120-day rolling beta of the IBEX with respect to the SPX is calculated. This shows how sensitive European equity risk is to U.S. market movements, and whether that sensitivity strengthens or weakens across different regimes. Both series are plotted to visually track changes in dependence, and the resulting metrics are merged back into the main analysis panel for use in later steps.


```python
# STEP 4 — ROLLING BETA & CORRELATION

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 4.1 Prepare return panel (IBEX, SPX)
print("Preparing return panel for rolling correlation & beta...")

# Tus retornos reales se llaman IBEX y SPX (no IBEX_ret / SPX_ret)
ret_panel = analysis_panel[["IBEX", "SPX"]].dropna()

print("Return panel shape:", ret_panel.shape)

# 4.2 Rolling correlation (IBEX ↔ SPX)
print("\nComputing rolling correlation (IBEX vs SPX)...")

window_corr = 60    # ~3 meses

ret_panel["rolling_corr"] = (
    ret_panel["IBEX"].rolling(window_corr).corr(ret_panel["SPX"])
)

print("Rolling correlation computed:", ret_panel["rolling_corr"].dropna().shape)

# Plot rolling correlation
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(ret_panel.index, ret_panel["rolling_corr"], linewidth=1.2)
ax.axhline(0, color="black", linestyle="--", linewidth=1)

ax.set_title("Rolling Correlation — IBEX vs SPX (60-day window)")
ax.set_ylabel("Correlation")
ax.set_xlabel("Date")
ax.grid(True)

plt.tight_layout()

# SAVE FIGURE 1
fig.savefig(RISK8_DIR / "06_rolling_correlation.png", dpi=300, bbox_inches="tight")

plt.show()

# 4.3 Rolling Beta (IBEX | SPX)
print("\nComputing rolling beta (IBEX | SPX)...")

window_beta = 120    # ~6 meses

def compute_beta(y, x):
    """
    Computes beta = Cov(y,x) / Var(x).
    y = IBEX, x = SPX
    """
    cov = np.cov(y, x, ddof=1)[0, 1]
    var = np.var(x, ddof=1)
    return cov / var if var != 0 else np.nan

betas = []
idx = ret_panel.index

for i in range(window_beta, len(ret_panel)):
    y = ret_panel["IBEX"].iloc[i - window_beta:i]
    x = ret_panel["SPX"].iloc[i - window_beta:i]
    betas.append(compute_beta(y.values, x.values))

beta_series = pd.Series(betas, index=idx[window_beta:])
ret_panel["rolling_beta"] = beta_series

print("Rolling beta computed:", ret_panel["rolling_beta"].dropna().shape)

# 4.4 Plot rolling beta
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(ret_panel.index, ret_panel["rolling_beta"], linewidth=1.2)
ax.axhline(1.0, color="black", linestyle="--", linewidth=1)

ax.set_title("Rolling Beta (IBEX | SPX) — 120-day window")
ax.set_ylabel("Beta coefficient")
ax.set_xlabel("Date")
ax.grid(True)

plt.tight_layout()
#Save figure 
fig.savefig(RISK8_DIR / "07_rolling_beta.png", dpi=300, bbox_inches="tight")

plt.show()

# 4.5 Merge results back into analysis panel
analysis_panel.loc[ret_panel.index, "rolling_corr"] = ret_panel["rolling_corr"]
analysis_panel.loc[ret_panel.index, "rolling_beta"] = ret_panel["rolling_beta"]

print("\nStep 4 completed.")
print("Variables added to analysis_panel: rolling_corr, rolling_beta.")
```

### Step 5: Tail Risk Comparison (VaR / ES)

This step evaluates and compares the tail-risk behavior of the IBEX 35 and the S&P 500 using daily VaR and ES estimates. The VaR/ES results are first loaded and aligned with the original return series. Once aligned, the notebook computes VaR95 and VaR99 exceedances for each index by checking how often daily returns fall below the corresponding VaR thresholds. Exceedance severity—how far returns fall beyond the VaR cutoff—is also calculated to assess the magnitude of tail events.

A summary table reports the exceedance rates and average severity for both markets. The notebook then visualizes the rolling exceedance frequency (200-day window), showing periods where either index experiences unusually high tail risk. A professional scatter plot of exceedance severity highlights the intensity and timing of extreme losses.

These metrics and plots provide a direct comparison of EU vs US tail risk, indicating which market experiences more frequent and more severe violations of risk forecasts, and how this evolves across market regimes.


```python
# STEP 5 — TAIL RISK COMPARISON (VaR / ES) — CLEAN VERSION

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Starting Tail Risk Comparison (VaR/ES)...")

# 5.1 Load DAILY VaR/ES results
var_es_file = RESULTS_PATH / "var_es_results_daily.csv"

if var_es_file.exists():
    var_es = pd.read_csv(var_es_file, parse_dates=["Date"])
    var_es = var_es.sort_values(["Date", "Index"])
    print("Daily VaR/ES dataset loaded:", var_es.shape)
else:
    print("WARNING: var_es_results_daily.csv NOT FOUND.")
    var_es = None

# 5.2 Prepare returns (from original returns DF)
#      returns: columnas ["IBEX", "SPX"] en DECIMALES
rets = returns[["IBEX", "SPX"]].dropna()
rets.columns = ["IBEX_ret", "SPX_ret"]
print("Returns aligned:", rets.shape)

# 5.3 Compute exceedances for VaR95 / VaR99
if var_es is not None:
    # Pivot diario: VaR95_IBEX, VaR95_SPX, ...
    var_pivot = var_es.pivot(index="Date", columns="Index",
                             values=["VaR95", "VaR99"])
    var_pivot.columns = ["_".join(col) for col in var_pivot.columns]

    # Alinear fechas de VaR con las de los retornos
    var_pivot = var_pivot.reindex(rets.index)

    aligned = rets.join(var_pivot, how="inner")
    print("Merged panel:", aligned.shape)

    # VaR está en %, retornos en decimales → pasamos retornos a %
    aligned["IBEX_ret_pct"] = 100 * aligned["IBEX_ret"]
    aligned["SPX_ret_pct"]  = 100 * aligned["SPX_ret"]

    # Exceedances 95 / 99
    aligned["IBEX_exc_95"] = (aligned["IBEX_ret_pct"] < aligned["VaR95_IBEX"]).astype(int)
    aligned["SPX_exc_95"]  = (aligned["SPX_ret_pct"]  < aligned["VaR95_SPX"]).astype(int)

    aligned["IBEX_exc_99"] = (aligned["IBEX_ret_pct"] < aligned["VaR99_IBEX"]).astype(int)
    aligned["SPX_exc_99"]  = (aligned["SPX_ret_pct"]  < aligned["VaR99_SPX"]).astype(int)

    # Severity (también en %)
    aligned["IBEX_sev_95"] = np.where(
        aligned["IBEX_exc_95"] == 1,
        aligned["IBEX_ret_pct"] - aligned["VaR95_IBEX"],
        np.nan
    )
    aligned["SPX_sev_95"] = np.where(
        aligned["SPX_exc_95"] == 1,
        aligned["SPX_ret_pct"] - aligned["VaR95_SPX"],
        np.nan
    )

else:
    aligned = rets.copy()
    print("Skipping VaR/ES calculations.")

# 5.4 Summary table
if var_es is not None:
    summary = pd.DataFrame({
        "Index": ["IBEX", "SPX"],
        "Exc_rate_95": [aligned["IBEX_exc_95"].mean(), aligned["SPX_exc_95"].mean()],
        "Exc_rate_99": [aligned["IBEX_exc_99"].mean(), aligned["SPX_exc_99"].mean()],
        "Avg_sev_95":  [aligned["IBEX_sev_95"].mean(), aligned["SPX_sev_95"].mean()]
    })
    print("\nTail Risk Summary:")
    print(summary)

# 5.5 Rolling Exceedance Rate (VaR95)
if var_es is not None:
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(aligned["IBEX_exc_95"].rolling(200).mean(), label="IBEX (95%)", linewidth=1.2)
    ax.plot(aligned["SPX_exc_95"].rolling(200).mean(), label="SPX (95%)", linewidth=1.2)

    ax.axhline(0.05, color="black", linestyle="--")
    ax.set_title("Rolling VaR95 Exceedance Rate (200 days)")
    ax.set_ylabel("Exceedance frequency")
    ax.grid(True)
    ax.legend()
    plt.tight_layout()
    #save figure 
    fig.savefig(RISK8_DIR / "08_exceedance_rate_95.png", dpi=300, bbox_inches="tight")
    plt.show()

# 5.6 Scatter Plot — VaR95 Exceedance Severity 
if var_es is not None:

    fig, ax = plt.subplots(figsize=(12, 4))

    ax.scatter(
        aligned.index,
        aligned["IBEX_sev_95"],
        s=12,
        alpha=0.7,
        label="IBEX Severity (95%)"
    )

    ax.scatter(
        aligned.index,
        aligned["SPX_sev_95"],
        s=12,
        alpha=0.7,
        label="SPX Severity (95%)"
    )

    ax.axhline(0, color="black", linewidth=1)
    ax.set_title("VaR95 Exceedance Severity — IBEX vs SPX")
    ax.set_ylabel("Severity (return_x100 − VaR95)")
    ax.grid(True)
    ax.legend()
    plt.tight_layout()

#Save figure
    fig.savefig(RISK8_DIR / "09_exceedance_severity_95.png", dpi=300, bbox_inches="tight")
    plt.show()


# 5.7 Add variables to analysis_panel
if var_es is not None:
    for col in [
        "IBEX_exc_95", "SPX_exc_95",
        "IBEX_exc_99", "SPX_exc_99",
        "IBEX_sev_95", "SPX_sev_95"
    ]:
        analysis_panel[col] = aligned[col]

print("\nStep 5 completed (clean version).")
```

### Step 7: Analytical Summary (Economic Interpretation)

The combined analysis of volatility, drawdowns, dependence on the U.S. market, and tail-risk behavior leads to a clear and nuanced conclusion about the structural differences between European equities (represented by the IBEX 35) and U.S. equities (represented by the S&P 500). Throughout the entire sample, **the European market behaves as a structurally riskier system**, more exposed to deep losses, more sensitive to global shocks, and more prone to extreme events, while the U.S. market displays a more stable, diversified, and resilient risk profile.

Beginning with volatility, both realized and model-implied measures show that the IBEX operates with **a consistently higher baseline level of risk**. The IBEX/SPX volatility ratio remains above one for most of the period, indicating that European equities magnify the influence of the global cycle more than U.S. equities. This pattern reflects a market structure that is more cyclical, less diversified across global sectors, and more exposed to external shocks. Moreover, European volatility reacts more sharply during stress episodes, displaying a **procyclical risk pattern**: when global conditions deteriorate, volatility in the IBEX accelerates faster and reaches higher levels.

The drawdown analysis reinforces this structural vulnerability. Although both markets suffer severe losses during global crises, **European drawdowns tend to be deeper and more persistent**. The maximum historical drawdown of the IBEX is significantly worse than that of the S&P 500, and the recovery process is slower, less consistent, and more volatile. This suggests that long-term investors in European equities face a higher probability of prolonged underwater periods and more severe capital impairment during downturns, especially when compared to the more robust recovery tendencies of the U.S. market.

Dependence measures highlight an important asymmetry in cross-market dynamics. Rolling correlation shows strong co-movement between Europe and the U.S., particularly during global stress, but rolling beta provides the key insight: the IBEX frequently behaves as a **leveraged response to U.S. market shocks**, amplifying their magnitude. This asymmetry means that European equities offer limited diversification benefits against U.S. equities precisely when diversification would be most valuable — in crisis periods. In practical terms, the European market acts as a follower to U.S. market dynamics, absorbing U.S. shocks with greater intensity.

Tail-risk diagnostics add another layer to this interpretation. The IBEX exhibits **more frequent VaR violations** and slightly higher average severity during exceedance events. These violations tend to cluster more densely, reflecting heavier tails and more discontinuous return behavior. This pattern is consistent with sharper market reactions to macroeconomic announcements, regional uncertainty, or liquidity disruptions. While the S&P 500 also experiences tail events, its exceedance profile is more stable and aligns more closely with risk-model expectations, reinforcing its perception as a more orderly and liquid market.

Overall, the analysis converges on a single conclusion:  

The U.S. equity market offers a more stable, predictable, and resilient risk environment, while the European market presents a higher, more reactive, and more fragile risk profile.

For asset-allocation purposes, this implies that European equity exposure should be treated as a higher-risk satellite position, managed with tighter controls, smaller weights, or hedging overlays, particularly during periods of global instability. In contrast, U.S. exposures function more naturally as a core allocation, supported by deeper liquidity, stronger sectoral diversification, smoother recovery patterns, and a more consistent tail-risk structure.
