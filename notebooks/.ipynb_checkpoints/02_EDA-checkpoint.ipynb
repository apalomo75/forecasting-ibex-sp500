## 1. Data Loading & Setup  
In this first step, we import the cleaned daily closing prices for both indices from the `/data/interim/` folder.  
The idea is to verify that the datasets were correctly prepared in the previous notebook (`01_data_cleaning.ipynb`) and that they share a common business-day calendar.  

```python
# 1. DATA LOADING & SETUP
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import sys

# Export module (for sending results to Power BI)
ROOT = Path.cwd().parent
sys.path.append(str(ROOT / "src"))
from export import export

# Correct paths
DATA = Path("../data/interim")
EDA_DIR = Path("../figures/02_EDA")
EDA_DIR.mkdir(parents=True, exist_ok=True)

# Load datasets
ibex = pd.read_csv(DATA / "IBEX_clean.csv", parse_dates=["Date"]).set_index("Date")
spx  = pd.read_csv(DATA / "SPX_clean.csv",  parse_dates=["Date"]).set_index("Date")

# Basic checks
print("IBEX shape:", ibex.shape)
print("SPX shape:", spx.shape)
display(ibex.head(3))
display(spx.head(3))
```

## 2. Descriptive Statistics  
Daily log-returns provide a scale-free measure of percentage change and are additive over time.  
Here we compute the logarithmic returns for both indices, store them for later modeling, and review their main descriptive statistics.

### Log-Returns 

To transform price series into stationary return series suitable for volatility modeling and forecasting,
we compute **logarithmic returns**, defined as:

$$
r_t = \log\left(\frac{P_t}{P_{t-1}}\right)
$$

Log-returns are preferred because they are time-additive, handle percentage changes naturally,
and stabilize variance across different price levels.



```python
# 2. COMPUTE LOG RETURNS
import numpy as np
from pathlib import Path

# Ensure numeric dtype (important if this cell runs independently)
ibex["IBEX"] = pd.to_numeric(ibex["IBEX"], errors="coerce")
spx["SPX"]   = pd.to_numeric(spx["SPX"], errors="coerce")

# Drop corrupted or missing rows
ibex = ibex.dropna()
spx  = spx.dropna()

# Compute log returns
returns = pd.DataFrame({
    "IBEX_ret": np.log(ibex["IBEX"] / ibex["IBEX"].shift(1)),
    "SPX_ret":  np.log(spx["SPX"] / spx["SPX"].shift(1))
}).dropna()

# Save for later modeling
FEATURES = Path("../data/features")
FEATURES.mkdir(parents=True, exist_ok=True)

returns.to_csv(FEATURES / "returns.csv", index=True)

# Quick look
returns.describe().T
```

```python
# Correlation between markets
corr = returns.corr().iloc[0, 1]

# Annualized volatility (sqrt(252) * daily std)
vol_ibex = returns["IBEX_ret"].std() * (252**0.5)
vol_spx  = returns["SPX_ret"].std() * (252**0.5)

print(f"IBEX annualized volatility: {vol_ibex:.2%}")
print(f"SPX annualized volatility:  {vol_spx:.2%}")
print(f"IBEX–SPX correlation:       {corr:.2f}")
```

This section explores the dynamic behaviour of the IBEX 35 and S&P 500 through graphical analysis.  

We’ll focus on four complementary visual analyses:

### 3.4 Rolling Correlation

To study how the comovement between IBEX 35 and S&P 500 evolves over time, we compute a **rolling (moving-window) correlation** that allows us to capture periods of increasing integration, divergence, or structural breaks.

The rolling correlation with ¨k¨ window size is defined as:

$$
\rho_{t}^{(k)} =
\frac{
\sum_{i=t-k+1}^{t}
(x_i - \bar{x})(y_i - \bar{y})
}{
\sqrt{
\sum_{i=t-k+1}^{t}(x_i - \bar{x})^2
}
\sqrt{
\sum_{i=t-k+1}^{t}(y_i - \bar{y})^2
}
}
$$

```python
# 3.4 CORRELATION DYNAMICS

window = 60  # 60-day rolling window
rolling_corr = returns["IBEX_ret"].rolling(window).corr(returns["SPX_ret"])

fig, ax = plt.subplots(figsize=(12, 5))
rolling_corr.plot(ax=ax, color="tab:blue", linewidth=1)
ax.set_title(f"Rolling {window}-Day Correlation — IBEX vs S&P 500")
ax.set_ylabel("Correlation")
ax.set_ylim(-1, 1)
plt.axhline(0, color="gray", linestyle="--", linewidth=0.8)
plt.tight_layout()
plt.show()

# SAVE FIGURE
fig.savefig(EDA_DIR / "eda_correlation.png", dpi=300, bbox_inches="tight")
```

## 4. Statistical Testing  

This section validates the statistical properties observed during the visual exploration through econometric tests. While plots revealed potential non-stationarity, volatility clustering, and non-normality, here we test these features using three standard diagnostics.

The objective is to confirm whether the data meet the assumptions required for time-series modeling and to determine which models (ARIMA, GARCH, etc.) are appropriate for the next stages.

We’ll perform the following tests:

1. **ADF (Augmented Dickey–Fuller)** tests for stationarity of the return series.  
2. **ARCH–LM (Engle’s Lagrange Multiplier)** tests for conditional heteroskedasticity (ARCH effects).  
3. **Jarque–Bera** tests whether the distribution of returns departs from normality.  

Combined, these tests offer statistical proof of the dynamics of both indices' means and variances.

```python
from statsmodels.tsa.stattools import adfuller

# 4.1 ADF 
def adf_test(series, name="Series"):
    result = adfuller(series, autolag="AIC")
    
    print(f"\nADF Test: {name}")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value:        {result[1]:.4f}")
    
    for key, value in result[4].items():
        print(f"Critical Value {key}: {value:.4f}")
    
    print() 

# Run test
adf_test(returns["IBEX_ret"], "IBEX Returns")
adf_test(returns["SPX_ret"], "S&P 500 Returns")

```

```python
from statsmodels.stats.diagnostic import het_arch

# 4.2 ARCH–LM
def arch_lm_test(series, name="Series", lags=12):
    lm_stat, lm_pval, f_stat, f_pval = het_arch(series, nlags=lags)
    
    print(f"\nARCH–LM Test: {name}")
    print(f"LM Statistic: {lm_stat:.4f}")
    print(f"p-value (LM): {lm_pval:.4f}")
    print(f"F Statistic:  {f_stat:.4f}")
    print(f"p-value (F):  {f_pval:.4f}")
    print()

# Run test
arch_lm_test(returns["IBEX_ret"], "IBEX Returns")
arch_lm_test(returns["SPX_ret"], "S&P 500 Returns")

```

```python
from scipy.stats import jarque_bera

# 4.3 JARQUE–BERA
def jarque_bera_test(series, name="Series"):
    jb_stat, jb_pval = jarque_bera(series)
    
    print(f"\nJarque–Bera Test: {name}")
    print(f"JB Statistic: {jb_stat:.4f}")
    print(f"p-value:       {jb_pval:.4f}")
    print()

# Run test
jarque_bera_test(returns["IBEX_ret"], "IBEX Returns")
jarque_bera_test(returns["SPX_ret"], "S&P 500 Returns")
```

### 4.4 Summary of Statistical Tests
The statistical diagnostics applied in this section validate the key assumptions of financial time series analysis.

**4.1 ADF (Augmented Dickey–Fuller Test):**
It has been chosen because it confirms that both the IBEX and S&P 500 return series are stationary, meaning their statistical properties —mean and variance— remain stable over time.
A p-value below 0.05 indicates rejection of the null hypothesis of a unit root, implying that the series does not follow a random walk and lacks a stochastic trend.
In other words, shocks to the series are temporary, and values tend to revert toward a long-term mean, making the data suitable for econometric modelling.
It is important to figure out because some predictive and volatility models, such as ARIMA or GARCH assume that the return process is stationary.

**4.2 ARCH–LM (Engle’s Lagrange Multiplier Test):** 
The use of ARCH–LM is to test whether the error variance changes over time.
A p-value below 0.05 indicates conditional heteroscedasticity, a defining feature of financial return series.
This means that return variance is not constant and that ARCH effects are present — periods of high volatility tend to cluster together.
Such behaviour violates the constant-variance assumption of classical models and supports the use of conditional variance models such as GARCH, or EGARCH, which can capture time-varying volatility.

**4.3 Jarque–Bera Test:** 
Here, the Jarque–Bera has been performed in order to test shows that return distributions deviate significantly from normality, exhibiting fat tails (high kurtosis) and asymmetry (skewness).
Statistically, this indicates that extreme market movements occur more frequently than predicted by a normal distribution, so subsequent models should incorporate non-Gaussian innovations, such as the Student’s t or Generalised Error Distribution, to better represent tail risk and heavy-tailed behaviour.

To sum up, the results confirm that return series are stationary, heteroskedastic, and non-normal, because of this, the next modelling stages must include conditional variance models with heavy-tailed distributions in order to capture the empirical dynamics of financial markets.

## 5. Analytical Insights

The exploratory and statistical analyses jointly describe the structural behaviour of the IBEX 35 and S&P 500 indices over the 2000–2025 period, confirming that both markets exhibit the typical properties of financial return series: stationarity, conditional heteroskedasticity, and non-normality.

1. Market Structure and Dynamics
Price trajectories show long-term growth punctuated by volatility spikes during major global crises (2008, 2011, 2020, 2022).
The IBEX 35 displays higher annualised volatility (≈22%) and slower post-crisis recoveries, while the S&P 500 maintains lower volatility (≈19%) and stronger resilience, reflecting greater depth and diversification. Despite these structural differences, both indices exhibit synchronous stress patterns and similar cyclical behaviour.

2. Return Distribution and Volatility Behaviour
Daily returns fluctuate around zero with volatility clustering, confirming the ARCH effects identified in statistical tests.
Calm periods alternate with sharp turbulence, indicating the necessity of variance models. Also, the presence of fat tails increases the probability of extreme movements, so we require heavy-tailed innovation distributions (e.g., Student-t) in the following modelling.

3. Cross-Market Correlation
The average correlation between daily returns is approximately 0.50, signalling moderate co-movement. It is important to remark that this dependence intensifies during crisis periods ( because of the contagion effect) and weakens in stable phases.

4. Modelling Implications
This supports a two-layer econometric modelling framework:

- The mean equation, return dynamics,  will be estimated through ARIMA and Prophet models to evaluate short-term predictability.
- The variance equation, risk dynamics,  will be modelled using EGARCH, capturing asymmetry and leverage effects.
Derived from these conditional volatility estimates, Value at Risk (VaR) and Expected Shortfall (ES) will provide quantitative measures of downside risk.

In conclusion, both markets demonstrate non-constant volatility, heavy-tailed distributions, and moderate correlation—empirical features that justify the use of modern econometric models capable of capturing non-linear and time-dependent risk dynamics.
