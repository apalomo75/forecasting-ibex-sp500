## 2. ARIMA Modeling 

To model short-term dependencies in financial returns, we use an **ARIMA(p,d,q)** specification. 

$$
\phi(L)\,(1 - L)^d\, y_t = \theta(L)\,\varepsilon_t
$$

This formulation allows ARIMA to capture short-memory patterns in returns, making it a natural benchmark for forecasting the IBEX 35 and S&P 500 under a parsimonious and statistically grounded framework.

This section focuses on fitting ARIMA models to the daily returns of the IBEX 35 and S&P 500 indices.  
We identify the optimal model orders (p, d, q) using information criteria, estimate the parameters, validate residuals with diagnostic tests, and generate short-term forecasts (1-day and 5-day).  

```python
from src import data_utils

# Load daily return data
returns = data_utils.load_returns_data()

print("Data loaded successfully:")
print(returns.head(), "\n")
print(f"Columns: {list(returns.columns)}")
```

```python
# GLOBAL SETUP — Load Clean Return Data
import sys
from pathlib import Path

# Define ROOT to the project root
ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

# Import the project package
from src import data_utils

# Load daily return data
returns = data_utils.load_returns_data()

print("Data loaded successfully:")
print(returns.head(), "\n")
print(f"Columns: {list(returns.columns)}")
```

```python
# 2. ARIMA MODELING 

# Imports
import warnings
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Project modules
from src.models import estimate_arima, forecast_arima, validate_residuals
from src.visualization import plot_acf_pacf, plot_forecast

# Enable inline graphics for Jupyter
%matplotlib inline

# Setup directories and suppress warnings

# Define folders
FIG_DIR = ROOT / "figures" / "03.ARIMA"
MODEL_DIR = ROOT / "data" / "models"
FIG_DIR.mkdir(parents=True, exist_ok=True)
MODEL_DIR.mkdir(parents=True, exist_ok=True)

# Enforce business-day frequency to avoid warnings
returns = returns.asfreq("B")

# Suppress all frequency-related and harmless runtime warnings
warnings.filterwarnings("ignore", message="No frequency information")
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)

print(f"Figures will be saved in: {FIG_DIR}")
print(f"Models will be saved in: {MODEL_DIR}\n")

# 1. Identify autocorrelation patterns (ACF / PACF)
print("Step 1 — ACF and PACF diagnostics")

fig1 = plot_acf_pacf(
    series=returns["IBEX_ret"].dropna(),
    lags=40,
    title="IBEX 35 — ACF and PACF",
    save_path=FIG_DIR / "arima_acf_pacf_ibex.png"
)
display(fig1)

fig2 = plot_acf_pacf(
    series=returns["SPX_ret"].dropna(),
    lags=40,
    title="S&P 500 — ACF and PACF",
    save_path=FIG_DIR / "arima_acf_pacf_spx.png"
)
display(fig2)

# 2. Estimate ARIMA models
print("\nStep 2 — Estimating optimal ARIMA models (AIC criterion)\n")

arima_ibex = estimate_arima(
    series=returns["IBEX_ret"],
    max_p=2, max_q=2, d=0,
    criterion="aic",
    label="IBEX 35"
)

arima_spx = estimate_arima(
    series=returns["SPX_ret"],
    max_p=2, max_q=2, d=0,
    criterion="aic",
    label="S&P 500"
)

print("IBEX 35 ARIMA summary")
display(pd.DataFrame(arima_ibex.summary().tables[1]))

print("\n S&P 500 ARIMA summary")
display(pd.DataFrame(arima_spx.summary().tables[1]))

# 3.Validate residuals
print("\nStep 3 — Residual diagnostics")

results_ibex = validate_residuals(arima_ibex, label="IBEX 35")
results_spx = validate_residuals(arima_spx, label="S&P 500")

validation_df = pd.concat([results_ibex, results_spx])
validation_path = MODEL_DIR / "arima_residual_tests.csv"
validation_df.to_csv(validation_path, index=False)

print(f"Residual diagnostics saved → {validation_path}")
display(validation_df)

# 4. Forecast (1-day and 5-day ahead)
print("\nStep 4 — Forecasts")

forecast_ibex_arima = forecast_arima(model=arima_ibex, steps=[1, 5], label="IBEX 35")
forecast_spx_arima  = forecast_arima(model=arima_spx, steps=[1, 5], label="S&P 500")

print("\nForecast results:")
display(forecast_ibex_arima)
display(forecast_spx_arima)

# 5. Display forecast plots

fig3 = plot_forecast(
    forecast_ibex_arima,
    title="IBEX 35 — ARIMA Forecasts (1-day & 5-day)",
    save_path=FIG_DIR / "arima_forecast_ibex.png"
)
display(fig3)

fig4 = plot_forecast(
    forecast_spx_arima,
    title="S&P 500 — ARIMA Forecasts (1-day & 5-day)",
    save_path=FIG_DIR / "arima_forecast_spx.png"
)

display(fig4)

print("\n✓ ARIMA modeling complete — results and figures displayed.")
```

- Estimated ARIMA models for each index and validated the residuals through Ljung–Box and Jarque–Bera tests.

These results will serve as the baseline for later model comparison with Prophet. 
### Interpretation 

Unlike the exploratory analysis (EDA), where price levels exhibited strong and persistent autocorrelation,  
the **daily log-returns** of both (Ibex and SP) indices show correlations close to zero.

- **Price levels** are non-stationary and follow a random walk pattern, so their ACF remains high across lags.  
- **Returns**, although, are stationary and reflect market efficiency: past price changes carry little linear predictive power.  
- While linear autocorrelation vanishes, volatility clustering remains. 

Therefore, the near-zero autocorrelation here confirms the statistical properties of financial return series rather than indicating an issue.



## 3. Prophet Modeling 
Prophet is a decomposable forecasting model designed to capture the main components that drive the evolution of financial and economic time series. The model is fully additive and can flexibly adapt to structural changes, making it suitable for data with regime shifts, smooth drifts, or non-linear growth. The general Prophet formulation is:

$$
y(t) = g(t) + s(t) + h(t) + \varepsilon_t
$$

In this section, the Prophet model is applied to the **daily log-returns** of the IBEX 35 and S&P 500 indices.  
The goal is to estimate short-term forecasts using Prophet’s additive decomposition of trend and seasonality.  

Each model will be fitted on historical returns, then used to produce **1-day** and **5-day ahead** predictions and the resulting forecasts and residual diagnostics will later be compared against the ARIMA results to evaluate model robustness and predictive accuracy.


```python
# 3. PROPHET MODELING
# Imports
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from prophet import Prophet
from IPython.display import display

from src.tests import jb_test
from src.visualization import plot_forecast

# Enable inline plotting
%matplotlib inline

# Setup directories and suppress warnings
FIG_DIR = ROOT / "figures" / "03.PROPHET"
MODEL_DIR = ROOT / "data" / "models"
FIG_DIR.mkdir(parents=True, exist_ok=True)
MODEL_DIR.mkdir(parents=True, exist_ok=True)

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", message="FutureWarning")
warnings.filterwarnings("ignore", message="ConvergenceWarning")

print(f"Figures will be saved in: {FIG_DIR}")
print(f"Models  will be saved in: {MODEL_DIR}\n")

# 1. Prepare data for Prophet
print("Step 1 — Preparing data for Prophet format")

def prepare_prophet_df(series):
    """Convert pandas Series to Prophet-compatible DataFrame."""
    df = pd.DataFrame({
        "ds": series.index,
        "y": series.values
    })
    return df.dropna()

ibex_df = prepare_prophet_df(returns["IBEX_ret"])
spx_df  = prepare_prophet_df(returns["SPX_ret"])

display(ibex_df.head())
display(spx_df.head())

# 2. Fit Prophet models
print("\nStep 2 — Fitting Prophet models")

prophet_ibex = Prophet(daily_seasonality=False, weekly_seasonality=True, yearly_seasonality=True)
prophet_spx  = Prophet(daily_seasonality=False, weekly_seasonality=True, yearly_seasonality=True)

prophet_ibex.fit(ibex_df)
prophet_spx.fit(spx_df)

# 3. Forecast 1-day and 5-day ahead
print("\nStep 3 — Forecasting 1-day and 5-day ahead")

future_ibex = prophet_ibex.make_future_dataframe(periods=5, freq="B")
forecast_ibex_prophet = prophet_ibex.predict(future_ibex).tail(5)[["ds", "yhat"]].rename(columns={"yhat": "Forecast"})
forecast_ibex_prophet["Horizon"] = [1, 2, 3, 4, 5]

future_spx = prophet_spx.make_future_dataframe(periods=5, freq="B")
forecast_spx_prophet = prophet_spx.predict(future_spx).tail(5)[["ds", "yhat"]].rename(columns={"yhat": "Forecast"})
forecast_spx_prophet["Horizon"] = [1, 2, 3, 4, 5]

print("\nIBEX 35 Prophet forecasts:")
display(forecast_ibex_prophet)

print("\nS&P 500 Prophet forecasts:")
display(forecast_spx_prophet)

# 4. Residual diagnostics
print("\nStep 4 — Residual diagnostics")

def prophet_residuals(model, df):
    """Compute residuals for Prophet model."""
    forecast = model.predict(df)
    residuals = df["y"].values - forecast["yhat"].values
    return residuals

res_ibex = prophet_residuals(prophet_ibex, ibex_df)
res_spx  = prophet_residuals(prophet_spx, spx_df)

# Apply Jarque–Bera test (normality of residuals)
jb_ibex = jb_test(res_ibex)
jb_spx  = jb_test(res_spx)

diagnostics = pd.DataFrame([
    {"Index": "IBEX 35", "JB_stat": jb_ibex[0], "JB_pvalue": jb_ibex[1]},
    {"Index": "S&P 500", "JB_stat": jb_spx[0], "JB_pvalue": jb_spx[1]}
])

diag_path = MODEL_DIR / "prophet_diagnostics.csv"
diagnostics.to_csv(diag_path, index=False)

print(f"Residual diagnostics saved → {diag_path}")
display(diagnostics)

# 5. Plot forecasts and save figures 

print("Saving Prophet figures into:", FIG_DIR)

# IBEX
fig_ibex = prophet_ibex.plot(prophet_ibex.predict(future_ibex))
plt.title("IBEX 35 — Prophet Forecast")
plt.tight_layout()

output_ibex = FIG_DIR / "prophet_forecast_ibex.png"
fig_ibex.savefig(output_ibex, dpi=300, bbox_inches="tight")
print("Saved:", output_ibex, output_ibex.exists())

plt.show()  # Show in notebook
plt.close(fig_ibex)

# SPX
fig_spx = prophet_spx.plot(prophet_spx.predict(future_spx))
plt.title("S&P 500 — Prophet Forecast")
plt.tight_layout()

output_spx = FIG_DIR / "prophet_forecast_spx.png"
fig_spx.savefig(output_spx, dpi=300, bbox_inches="tight")
print("Saved:", output_spx, output_spx.exists())

plt.show()  # Show in notebook
plt.close(fig_spx)

print("\n Prophet figures saved and displayed.")



```

# Technical Summary — Prophet Results
Residual diagnostics show high Jarque–Bera statistics (p < 0.05), indicating non-normal residuals.  
This behavior is expected as we show in the previous notebook (EDA), where we saw that heavy tails and volatility clustering dominate.

While ARIMA focuses on linear dependencies, Prophet offers a complementary non-linear framework capable of capturing structural breaks or regime changes, which is why we will later make a comparison between the results.

### Interpretation 

The Prophet model provided smooth and stable forecasts for both indices, with predicted daily returns oscillating near zero, consistent with market efficiency expectations.  

However, residual diagnostics (Jarque–Bera test) reveal strong deviations from normality, confirming that the model fails to capture the heavy tails and volatility clustering typical of financial returns.  

This outcome is an inherent limitation of additive trend models when applied to high-frequency financial data.  
Prophet successfully captures non-linear trends and potential regime changes,  but the conditional variance dynamics remain heteroscedastic.  

These differences will be quantitatively compared in the next section and subsequently modeling the variance with **EGARCH** to handle volatility effects.


```python
import numpy as np

# Step 1 — Compute performance metrics (MAE & RMSE)

print("\nStep 1 — Computing performance metrics (MAE, RMSE)")

def compute_metrics(actual, predicted, model_name, label):
    mae = np.mean(np.abs(actual - predicted))
    rmse = np.sqrt(np.mean((actual - predicted)**2))
    return {"Model": model_name, "Index": label, "MAE": mae, "RMSE": rmse}

# Filter Prophet forecasts for horizons 1 and 5
forecast_ibex_prophet_filtered = forecast_ibex_prophet[forecast_ibex_prophet["Horizon"].isin([1, 5])]
forecast_spx_prophet_filtered  = forecast_spx_prophet[forecast_spx_prophet["Horizon"].isin([1, 5])]

# Actual returns for comparison (last + last-5)
actual_ibex = np.array([returns["IBEX_ret"].iloc[-1], returns["IBEX_ret"].iloc[-5]])
actual_spx  = np.array([returns["SPX_ret"].iloc[-1], returns["SPX_ret"].iloc[-5]])

# Compute metrics
metrics = []
metrics.append(compute_metrics(actual_ibex, forecast_ibex_arima["Forecast"].values, "ARIMA", "IBEX 35"))
metrics.append(compute_metrics(actual_ibex, forecast_ibex_prophet_filtered["Forecast"].values, "Prophet", "IBEX 35"))

metrics.append(compute_metrics(actual_spx, forecast_spx_arima["Forecast"].values, "ARIMA", "S&P 500"))
metrics.append(compute_metrics(actual_spx, forecast_spx_prophet_filtered["Forecast"].values, "Prophet", "S&P 500"))

metrics_df = pd.DataFrame(metrics)
display(metrics_df)

# Step 2 — Diebold–Mariano test

print("\nStep 2 — Diebold–Mariano Test (ARIMA vs Prophet)")

def diebold_mariano(e1, e2):
    d = (e1**2) - (e2**2)
    mean_d = np.mean(d)
    var_d = np.var(d, ddof=1)
    dm_stat = mean_d / np.sqrt(var_d / len(d))
    from scipy.stats import t
    p_value = 2 * (1 - t.cdf(abs(dm_stat), df=len(d) - 1))
    return dm_stat, p_value

errors_ibex_arima    = actual_ibex - forecast_ibex_arima["Forecast"].values
errors_ibex_prophet  = actual_ibex - forecast_ibex_prophet_filtered["Forecast"].values

errors_spx_arima     = actual_spx - forecast_spx_arima["Forecast"].values
errors_spx_prophet   = actual_spx - forecast_spx_prophet_filtered["Forecast"].values

dm_ibex = diebold_mariano(errors_ibex_arima, errors_ibex_prophet)
dm_spx  = diebold_mariano(errors_spx_arima, errors_spx_prophet)

dm_results = pd.DataFrame([
    {"Index": "IBEX 35", "DM_stat": dm_ibex[0], "p_value": dm_ibex[1]},
    {"Index": "S&P 500", "DM_stat": dm_spx[0], "p_value": dm_spx[1]}
])

display(dm_results)

# Step 3 — Visualization of forecast comparison

print("\nStep 3 — Visualization of forecast comparison")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].bar(["ARIMA", "Prophet"], metrics_df.query("Index == 'IBEX 35'")["RMSE"])
axes[0].set_title("IBEX 35 — RMSE Comparison")
axes[0].set_ylabel("RMSE")

axes[1].bar(["ARIMA", "Prophet"], metrics_df.query("Index == 'S&P 500'")["RMSE"])
axes[1].set_title("S&P 500 — RMSE Comparison")
axes[1].set_ylabel("RMSE")

plt.tight_layout()
plt.show()

# Save figure to ARIMA and PROPHET folders
output_arima   = ROOT / "figures" / "03.ARIMA"   / "rmse_comparison.png"
output_prophet = ROOT / "figures" / "03.PROPHET" / "rmse_comparison.png"

fig.savefig(output_arima, dpi=300, bbox_inches="tight")
fig.savefig(output_prophet, dpi=300, bbox_inches="tight")

print("Saved in:")
print("-", output_arima)
print("-", output_prophet)

# Step 4 — Save metrics
results_path = MODEL_DIR / "arima_prophet_metrics.csv"
metrics_df.to_csv(results_path, index=False)
print(f"\nComparison metrics saved → {results_path}")

print("\n Model comparison complete — metrics, Diebold–Mariano test and visualizations displayed.")
```

###  ARIMA vs Prophet Comparison

Both approaches achieved very similar accuracy, with ARIMA showing slightly lower errors across both metrics.

The Mean Absolute Error (MAE) measures the average magnitude of the forecast errors, while the Root Mean Squared Error (RMSE) penalises larger deviations more heavily, making it more sensitive to volatility spikes and outliers.
ARIMA slightly outperforms Prophet in the IBEX 35, whereas Prophet achieves marginally better RMSE in the S&P 500.
However, the numerical differences in both markets are very small, below 0.001.

The Diebold–Mariano (DM) test, which statistically compares forecast accuracy, reports high p-values (p ≈ 0.65 for IBEX and p ≈ 0.61 for S&P 500).
Because these values are far above the 0.05 threshold, we reject the null hypothesis of equal forecast accuracy.
This indicates that both ARIMA and Prophet yield statistically equivalent predictive performance at the daily frequency. Since financial returns behave as weakly stationary and near-white-noise processes, both models offer limited predictive power.

These results reinforce that predictability in returns is minimal, in line with the Efficient Market Hypothesis. The remaining structure lies in the volatility, not in the mean, motivating the next phase: conditional variance modelling (EGARCH).


## 5. Analytical Insights
This notebook explored the short-term return dynamics of the IBEX 35 and S&P 500, testing the presence of linear predictability through **ARIMA** and **Prophet** models.

The workflow followed this sequence:

1. **Stationarity verification (ADF Test)** confirmed that the return series were stationary.  
2. **ARIMA modeling** captured short-term autocorrelation but produced residuals consistent with white noise, confirming that most linear dependence had been extracted.  
3. **Prophet modeling** offered smooth trend-based forecasts but failed to improve accuracy over ARIMA, reflecting the absence of deterministic seasonal or structural patterns at the daily frequency.  
4. **Model comparison** showed nearly identical predictive performance.

Overall, the findings reinforce the fact that **daily log-returns are weakly stationary, exhibit low predictability, and show evidence of volatility clustering**. Such behaviour cannot be captured by ARIMA or Prophet, whose assumptions rely on constant variance. Therefore, the next step in the modelling process is to move from the mean equation to the variance equation, specifically by estimating **EGARCH models**, capable of representing:

- **Conditional heteroscedasticity** (time-varying volatility)  
- **Leverage effects** (asymmetric volatility response to negative shocks)  
- **Fat-tailed residual distributions** (Student-t)
